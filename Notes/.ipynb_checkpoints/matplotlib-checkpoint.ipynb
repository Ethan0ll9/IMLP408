## matplotlib.pyplot
#### plot 方法繪出圖表
#### grid 方法顯示圖表的水平跟垂直格線

```python
import matplotlib.pyplot as plt
import pandas as pd

days = range(1, 8)
celsius = [25, 24.3, 23.5, 26, 23, 26, 28.5]
plt.plot(days, celsius, "r-o")
plt.xlabel("日期")
plt.ylabel("溫度")
plt.title("每日溫度表")
plt.grid(True)
plt.show()
```

## 繪製多條線於同一張圖
#### x軸標籤: xlabel()
#### y軸標籤: ylabel()
#### 標題: title()
#### 顯示每條線資訊: plot(label="")

```python
import numpy as np

x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, "r-o", label="sin(x)")
plt.plot(x, cosinus, "g--", label="cos(x)")
plt.xlabel("徑度")
plt.ylabel("振幅")
plt.title("sin 和 cos 波形")
plt.legend(loc=1)
plt.show()
```
