## matplotlib.pyplot
#### plot 方法繪出圖表
#### grid 方法顯示圖表的水平跟垂直格線

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
days = range(1, 8)
celsius = [25, 24.3, 23.5, 26, 23, 26, 28.5]
plt.plot(days, celsius, "r-o")
plt.xlabel("日期")
plt.ylabel("溫度")
plt.title("每日溫度表")
plt.grid(True)
plt.show()
```

## 繪製多條線於同一張圖
#### x軸標籤: xlabel()
#### y軸標籤: ylabel()
#### 標題: title()
#### 顯示每條線資訊: plot(label="")
#### axis 顯示軸的範圍

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, "r-o", label="sin(x)")
plt.plot(x, cosinus, "g--", label="cos(x)")
plt.xlabel("徑度")
plt.ylabel("振幅")
plt.title("sin 和 cos 波形")
plt.legend(loc=1)
print(plt.axis())
plt.show()
```

## 繪製sin()三角函數的散佈圖
#### scatter 方法繪製散佈圖，散佈圖為點的集合

```python
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
plt.scatter(x, y)
plt.show()
```

## 子圖表(Subplots)
#### plt.subplots(num_rows, num_columns, plot_nums)
#### plot_nums 參數是顯示第幾張圖表，其值是從1到最大儲存格數的num_rows * num_cols，繪製方向是先水平在垂直

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.subplot(2, 1, 1)
plt.plot(x, sinus, "r-o")
plt.subplot(2, 1, 2)
plt.plot(x, cosinus, "g--")
plt.show()
```
